```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
l1,l2,l3,l4,l5=[],[],[],[],[]
for page in range(1,15):
    link=f"https://www.jobberman.com/jobs/it-telecoms?page={page}"
    re=requests.get(link)
    data=BeautifulSoup(re.content,"lxml")
    role=data.find_all("p", class_="text-lg font-medium break-words text-link-500")
    job_function=data.find_all("p",class_="text-sm text-gray-500 text-loading-animate inline-block")
    company=data.find_all("a", class_="text-loading-animate text-loading-animate-link")
    details=data.find_all("p", class_="ml-auto text-sm font-normal text-gray-700 text-loading-animate")
    post=data.find_all("p", class_="text-sm font-normal text-gray-700 md:text-gray-500 md:pl-5")
    
    max_length=max(len(role),len(job_function),len(company),len(post),len(details))
    
    for i in range(max_length):
        if i >= len(role) or not role[i].text.strip():  
            continue 
        l1.append(role[i].text.strip()) 
        l2.append(job_function[i].text.replace("Job Function", "").replace(":", "").strip() if i < len(job_function) else "N/A")
        l3.append(company[i].text.strip() if i < len(company) else "N/A")
        l4.append(details[i].text.strip() if i < len(details) else "N/A")
        l5.append(post[i].text.strip() if i < len(post) else "N/A")
print("Scraping completed.")
```

```python
file = pd.DataFrame({"Job_Role" :l1 ,"Company_Name" : l3, 'Job_function':l2,"Description" : l5 , "Post_Time" : l4 })
file
```
